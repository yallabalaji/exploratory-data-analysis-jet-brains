```python
activities_df = total_df.filter(like='activities_kind')
activities_df.info()
```

```python
activities_df_group = activities_df.sum()
```

```python
activities_df_group = activities_df_group.to_frame()
```

```python
activities_df_group['Roles'] = activities_df_group['Roles'].str.replace('activities_kinds.', '')
activities_df_group = activities_df_group.sort_values('Total_Members',ascending=False)
activities_df_group
```

```python
import plotly.graph_objects as go


# Create the bar plot
fig = go.Figure(go.Bar(
    x=activities_df_group.Roles, y=activities_df_group.Total_Members,
    marker_color=px.colors.qualitative.Dark2
))

# Add title and axis labels
fig.update_layout(
    title='Tech Domain Demographics: A Comparative Analysis',
    xaxis_title='Roles',
    yaxis_title='Total Members'
)

# Show the plot
fig.show()
```
